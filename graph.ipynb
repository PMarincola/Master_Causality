## Age distribution

```python
age_distri_dt = pd.read_excel('./AgeDistribution.xls')
age_distri_dt.info()
```

```python
age_distri_dt.head(5)
```

```python
#Convert to array
#AGE array
age_df = age_distri_dt.iloc[0:16]
x_age = age_df.Age
x_age = pd.Series(x_age).array
#veteran array
y_vet = age_distri_dt.loc[age_distri_dt['Category'] == 'Veterans']
y_vet = y_vet.Distri
y_vet = pd.Series(y_vet).array
#Non veteran array
y_Nonvet = age_distri_dt.loc[age_distri_dt['Category'] == 'Nonveterans']
y_Nonvet = y_Nonvet.Distri
y_Nonvet = pd.Series(y_Nonvet).array
```

```python
#Barplot distribution
n= len(y_Nonvet)
r = np.arange(n)
width = 0.25

plt.bar(r, y_vet, color = 'b',
        width = width, edgecolor = 'black',
        label='Veterani')
plt.bar(r + width, y_Nonvet, color = 'g',
        width = width, edgecolor = 'black',
        label='Non Veterani')
  
plt.xlabel("Età")
plt.ylabel("Percentuale intervistati")
plt.title("Distribuzione età inetrvistati")

# plt.grid(linestyle='--')
plt.xticks(r + width/2,x_age)
plt.legend(fontsize=20)
```

## Total Income

```python
earnings_dt_totalIncome = pd.read_excel('./TotalIncome2018.xlsx')
earnings_dt_totalIncome.info()
```

```python
earnings_dt_totalIncome.dropna()
```

```python
#Assign values to variables
vet_Earnings = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Veterans') & 
                                            (earnings_dt_totalIncome['Type'] == 'Earnings')].USD
vet_Social_Security = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Veterans') & 
                                            (earnings_dt_totalIncome['Type'] == 'Social Security')].USD
vet_Investments = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Veterans') & 
                                            (earnings_dt_totalIncome['Type'] == 'Investments')].USD
vet_Retirement_Plans =earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Veterans') & 
                                            (earnings_dt_totalIncome['Type'] == 'Retirement Plans')].USD
vet_VA_Disabililty = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Veterans') & 
                                            (earnings_dt_totalIncome['Type'] == 'VA Disabililty')].USD
Nonvet_Earnings = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Nonveteran') & 
                                            (earnings_dt_totalIncome['Type'] == 'Earnings')].USD
Nonvet_Social_Security = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Nonveteran') & 
                                            (earnings_dt_totalIncome['Type'] == 'Social Security')].USD
Nonvet_Investments = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Nonveteran') & 
                                            (earnings_dt_totalIncome['Type'] == 'Investments')].USD
Nonvet_Retirement_Plans = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Nonveteran') & 
                                            (earnings_dt_totalIncome['Type'] == 'Retirement Plans')].USD
Nonvet_VA_Disabililty = earnings_dt_totalIncome[(earnings_dt_totalIncome['Category'] == 'Nonveteran') & 
                                            (earnings_dt_totalIncome['Type'] == 'VA Disabililty')].USD
```

```python
#array creation for graph
labels = ['Veterans','Nonveteran']
earnings = []
earnings = np.append(earnings,vet_Earnings)
earnings = np.append(earnings,Nonvet_Earnings)
Social_Security = []
Social_Security = np.append(Social_Security,vet_Social_Security)
Social_Security = np.append(Social_Security,Nonvet_Social_Security)
Investments = []
Investments = np.append(Investments,vet_Investments)
Investments = np.append(Investments,Nonvet_Investments)
Retirement_Plan = []
Retirement_Plan = np.append(Retirement_Plan,vet_Retirement_Plans)
Retirement_Plan = np.append(Retirement_Plan,Nonvet_Retirement_Plans)
VA_Disabililty = []
VA_Disabililty = np.append(VA_Disabililty,vet_VA_Disabililty)
VA_Disabililty = np.append(VA_Disabililty,Nonvet_VA_Disabililty)
```

```python
print(earnings),print(Social_Security),print(Investments),print(Retirement_Plan),print(VA_Disabililty)
```

```python
sns.set(rc={'figure.figsize':(15,6)})
width = 0.25 
fig, ax = plt.subplots()

ax.bar(labels, earnings, width, label='Earnings')
ax.bar(labels, Social_Security, width,label='Social_Security', bottom=earnings)
ax.bar(labels, Investments, width, label='Investments',bottom=earnings+Social_Security)
ax.bar(labels, Retirement_Plan, width,label='Retirement_Plans',bottom=earnings+Social_Security+Investments)
ax.bar(labels, VA_Disabililty, width,label='VA_Disabililty',bottom=earnings+Social_Security+Investments+Retirement_Plan)

plt.text(x = 0, # x-coordinate position of data label
y = 66000, # y-coordinate position of data label, adjusted to be 150 below the data point
s = '$65500', # data label
fontsize=20, 
color = 'black',ha='center')

plt.text(x = 1, # x-coordinate position of data label
y = 66000, # y-coordinate position of data label, adjusted to be 150 below the data point
s = '$65700', # data label
fontsize=20, 
color = 'black',ha='center')

ax.set_ylabel('Entrate USD')
ax.set_title('Entrate totali veterani e non veterani -71 anni - survey 2018')
ax.legend(fontsize=20)
plt.ylim((0.0,70000))
plt.show()
#bottom=vet_Earnings,
```
